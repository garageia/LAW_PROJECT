```python
import os
import csv
import pytesseract
from PIL import Image

# Chemin vers le dossier contenant les fichiers
dossier_data = 'C:/Users/lenovo/Desktop/jfk_project/data'

# Liste pour stocker les contenus extraits
contenus_extraits = []

# Parcours des fichiers dans le dossier
for fichier in os.listdir(dossier_data):
    chemin_fichier = os.path.join(dossier_data, fichier)
    extension = os.path.splitext(fichier)[1].lower()

    if extension == '.jpg' or extension == '.png':
        # Ouverture de l'image
        image = Image.open(chemin_fichier)

        # Extraction du contenu de l'image
        contenu_image = pytesseract.image_to_string(image, lang='fra')

        contenus_extraits.append(contenu_image)

# Enregistrement des contenus extraits dans un fichier CSV
chemin_csv = 'C:/Users/lenovo/Desktop/jfk_project/data.csv'

with open(chemin_csv, 'w', newline='', encoding='utf-8') as fichier_csv:
    writer = csv.writer(fichier_csv)
    for contenu in contenus_extraits:
        writer.writerow([contenu])
```

```python
import os
import csv
import pytesseract
from PIL import Image

# Chemin vers le dossier contenant les fichiers
dossier_data = 'C:/Users/lenovo/Desktop/jfk_project/data'

# Liste pour stocker les contenus extraits
contenus_extraits = []

# Parcours des fichiers dans le dossier
for fichier in os.listdir(dossier_data):
    chemin_fichier = os.path.join(dossier_data, fichier)
    extension = os.path.splitext(fichier)[1].lower()

    if extension == '.pdf':
        # Ignorer l'extraction des fichiers PDF pour le moment
        pass
    elif extension == '.jpg' or extension == '.png':
        # Extraction du contenu des fichiers JPG et PNG
        image = Image.open(chemin_fichier)
        contenu_image = pytesseract.image_to_string(image, lang='fra')
        contenus_extraits.append(contenu_image)

# Enregistrement des contenus extraits dans un fichier CSV
chemin_csv = 'C:/Users/lenovo/Desktop/jfk_project/data1.csv'

with open(chemin_csv, 'w', newline='', encoding='utf-8') as fichier_csv:
    writer = csv.writer(fichier_csv)
    for contenu in contenus_extraits:
        writer.writerow([contenu])
```
